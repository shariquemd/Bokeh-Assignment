```python
In Bokeh, glyphs are geometric shapes or visual markers that you can use to represent data on a plot. Common glyphs include lines, circles, squares, bars, and more. You can add glyphs to a Bokeh plot using methods provided by the figure object.
```

```python
from bokeh.plotting import figure, show

# Create a figure
p = figure(title="Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a line glyph
p.line(x=[1, 2, 3, 4, 5], y=[6, 7, 2, 4, 8], line_width=2, line_color="blue", legend_label="Line")

# Add a circle glyph
p.circle(x=[1, 2, 3, 4, 5], y=[6, 7, 2, 4, 8], size=10, color="red", legend_label="Circle")

# Add a square glyph
p.square(x=[1, 2, 3, 4, 5], y=[6, 7, 2, 4, 8], size=8, color="green", legend_label="Square")

# Show the legend
p.legend.location = "top_left"

# Show the plot
show(p)
```
